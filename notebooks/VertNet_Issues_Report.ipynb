```python
print "ISSUES updated since", since_date, ":"
for each in r.json():
    if "title" in each:
        print each["html_url"], each["title"]
        
```

```python
print "NEW issues since", since_date, ":"
for each in r.json():
    if "title" in each:
        for labels in each["labels"]:
            if "name" in labels:
                if labels["name"] == "new":
                    print each["html_url"], each["title"]
                    
```
