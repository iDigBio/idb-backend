```python
import requests
import idigbio

```

```python
api = idigbio.json()

```

```python
# https://search.idigbio.org/v2/search/records?rq={%22flags%22:%22dwc_basisofrecord_invalid%22}
record_list = api.search_records(rq={"flags": "dwc_basisofrecord_invalid"},fields="data.dwc:basisOfRecord",limit=600000)
```

```python
len(record_list["items"])
```

```python
# known bad values in basisOfRecord (we are looking for new good values)
dont_count_list=['Occurrence']

```

We will find mostly 'MaterialSample'.

```python
for each in record_list["items"]:
    if "dwc:basisOfRecord" in each["data"]:
        if each["data"]["dwc:basisOfRecord"] not in dont_count_list:
            print (each)
```

---

```python
# known bad values in basisOfRecord (we are looking for new good values)
dont_count_list=['Occurrence','MaterialSample']

```

Here are some of the other values providers have used.

```python
found_basisOfRecord_values = set()
for each in record_list["items"]:
    if "dwc:basisOfRecord" in each["data"]:
        if each["data"]["dwc:basisOfRecord"] not in dont_count_list:
            found_basisOfRecord_values.add(each["data"]["dwc:basisOfRecord"])

print(found_basisOfRecord_values)
            
```
