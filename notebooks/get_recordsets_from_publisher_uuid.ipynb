```python
import requests
```

```python
pub_uuid = '3b5899c7-f3af-45b7-b992-af5e98b39ecb'

url = 'http://search.idigbio.org/v2/search/recordsets/?rsq={%22publisher%22:%22' + pub_uuid + '%22}&fields=[%22name%22,%20%22publisher%22]&limit=1000'
```

```python
r = requests.get(url)
```

```python
r.content
```

```python
request_json = r.json()
```

```python
matched = {}
print (type(matched))
```

```python
request_json["items"]
```

```python
for item in request_json["items"]:
        u = item["uuid"]
        n = item["indexTerms"]["name"]
        matched[u] = n
        
```

```python
for each in matched:
    rs_url = 'https://www.idigbio.org/portal/recordsets/' + each
    print (matched[each])
    print (rs_url)
    print ('')
```
